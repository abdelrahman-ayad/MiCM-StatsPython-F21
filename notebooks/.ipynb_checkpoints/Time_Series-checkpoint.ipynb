### Time series analysis with pandas
### The following notebook has been inspired by the work in: https://towardsdatascience.com/machine-learning-part-19-time-series-and-autoregressive-integrated-moving-average-model-arima-c1005347b0d7


```python
# Import libraries and modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
fileName = 'https://raw.githubusercontent.com/abdelrahman-ayad/MiCM-StatsPython-F21/main/data/airline-passengers.csv'
df = pd.read_csv(fileName)
```

```python
df.head(10)
df.dtypes
```

```python
df.index = pd.to_datetime(df.index)
df.head()
df.describe()
```

```python
# Resample data
df["Magnitude"].resample("D").apply([np.mean]).plot()
plt.title("Magnitude average every two days")
plt.ylabel("Magnitude")
plt.show()
```

```python
df.plot(y="Magnitude", use_index = True)
plt.show()
```
